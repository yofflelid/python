```python
import pyautogui
import time
import math

time.sleep(3)  # 그림판으로 이동할 시간

center_x = 800
center_y = 1000
radius = 50

points = []

# 5개의 꼭짓점 계산
for i in range(5):
    angle = math.radians(90 + i * 72)  # 72도 간격
    x = center_x + radius * math.cos(angle)
    y = center_y - radius * math.sin(angle)
    points.append((x, y))

# 별 순서대로 연결 (0 → 2 → 4 → 1 → 3 → 0)
order = [0, 2, 4, 1, 3, 0]

pyautogui.moveTo(points[order[0]])
pyautogui.mouseDown()

for i in order[1:]:
    pyautogui.dragTo(points[i], duration=0.5)

pyautogui.mouseUp()
```

```python
import pyautogui
import math
import time

center_x = 1280
center_y = 800
radius = 300

points = []

for i in range(5):
    angle = math.radians(72 * i - 90)
    x = center_x + radius * math.cos(angle)
    y = center_y + radius * math.sin(angle)
    points.append((x, y))

time.sleep(3)

# 시작점 이동
pyautogui.moveTo(points[0])
pyautogui.mouseDown()

current = 0

for _ in range(5):
    next_point = (current + 2) % 5
    pyautogui.dragTo(points[next_point], duration=0.5)
    current = next_point

pyautogui.mouseUp()
```

```python
방향	값 변화
오른쪽	x 증가
왼쪽	x 감소
아래	y 증가
위	y 감소
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

print("=" *100)
print(" 이 크롤러는 대한민국구석구석 사이트의 자료 검색용 웹크롤러입니다.")
print("=" *100)

query_txt = input('1.수집할 자료의 키워드는 무엇입니까?(예: 제주도): ')
print("\n")

# 크롬 드라이버 실행
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

# 사이트 접속
url = 'https://korean.visitkorea.or.kr'
driver.get(url)
driver.maximize_window()

# ✅ Step 1. 돋보기 버튼 클릭
wait = WebDriverWait(driver, 10)

search_icon = wait.until(
    EC.element_to_be_clickable((By.CLASS_NAME, "btn_search"))
)

search_icon.click()

# ✅ Step 2. 검색창이 나타날 때까지 기다림
search_box = wait.until(
    EC.presence_of_element_located((By.ID, "inp_search_index"))
)

# ✅ Step 3. 검색어 입력
search_box.send_keys(query_txt)

# ✅ Step 4. 엔터 입력
search_box.send_keys(Keys.ENTER)

time.sleep(3)
```
