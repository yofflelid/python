```python
""""
Selenium의 .click()은 자식 요소를 클릭해도 부모 링크가 실행되는 경우가 많다.

현재 구조

<a>
   div
      div

이런 구조에서는 브라우저가 클릭 이벤트를 부모 a 태그로 전달한다.

이것을 이벤트 버블링(Event Bubbling) 이라고 한다.

즉 실제 동작은 이렇게 된다.

div 클릭
 → a 태그로 이벤트 전달
 → 링크 실행

그래서 다음이 모두 같은 결과를 만든다.

a 클릭
div 클릭
div 안의 div 클릭
"""
```

```python
# Step 9. 데이터 수집하기
no2=[]
title2=[]
date2=[]
contents2=[]


no = 1

for a in range(1, page_cnt+1):

    print("\n")
    print("%s 페이지 내용 수집 시작 =======================" % a)

    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')

    # 게시글 목록
    content_list = soup.find_all('div','rp_inlink_item')

    for i in content_list:

        try:
            # 제목
            title = i.find('a','rp_inlink').get_text().strip()

            # javascript 링크
            link = i.find('a','rp_inlink')['href']

        except:
            continue

        print("\n")
        print("%s 번째 정보를 추출하고 있습니다============" % no)

        # 제목 저장
        title2.append(title)
        print("1. 제목:",title)

        # 게시글 클릭 (javascript 실행)
        driver.execute_script(link)
        time.sleep(2)

        # 상세페이지 HTML
        html_1 = driver.page_source
        soup_1 = BeautifulSoup(html_1,'html.parser')

        # 공개일
        try:
            date = soup_1.find_all('span','sclinkage_evalue')[0].get_text().strip()
        except:
            date = "공개일 없음"

        date2.append(date)
        print("2. 공개일:",date)

        # 상담내용
        try:
            cont = soup_1.find('span','rp_indata').get_text("\n").strip()
        except:
            cont = "내용 없음"

        contents2.append(cont)
        print("3. 상담내역:",cont)

        # 목록으로 돌아가기
        driver.back()
        time.sleep(2)

        no += 1

        if no > cnt:
            break

    if no > cnt:
        break

    # 다음 페이지 이동
    if no > cnt:
        break

# 다음 페이지 이동
    if a < page_cnt:
        driver.execute_script("move(%s)" % (a+1))
        time.sleep(2)
print("요청하신 작업이 완료되었습니다")
```

```python
#Step 9. 데이터 수집하기
no2=[] # 게시글 번호 컬럼
title2=[ ] # 게시글 제목 컬럼
date2=[ ] # 게시글 날짜 컬럼
contents2=[] # 상담내용
full_url2=[] # 논문 원본 URL

no = 1 # 게시글 번호 초기값

for a in range(1,page_cnt+1) :
    print("\n")
    print("%s 페이지 내용 수집 시작합니다 =======================" %a)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    content_list = soup.find('div','rp_tb_wrap default_skin complaint_mayor_hope_list').find_all('div','rp_inlink_item')

    for i in content_list:
        # 여기서 바로 들어 가야겠네?? 아니면 제목 뽑고 들어가거나 ?? 

        # 민원신청 제목 체크하기
        try:
            title = soup_1.find('a','rp_inlink').get_text().strip()
        except :
            continue
        else :
            # 1.게시글 번호
            print("\n")
            print("%s 번째 정보를 추출하고 있습니다============" %no)
            no2.append(no)
            print("1.번호 : %s" %no)

            # 2. 민원신청 제목
            title2.append(title.strip())
            print("2.제목 : %s" %title.strip())

            # 3. 민원목록 안의 상세 내역에서 추출할 수 있음.
            url_1 = i.find('a`','rp_inlink_item').find('a')['href']
            full_url = 'https://eungdapso.seoul.go.kr/main.do'+url_1
            time.sleep(1)
            driver.get(full_url)

            html_1 = driver.page_source
            soup_1 = BeautifulSoup(html_1, 'html.parser')

            # 4. 공개일 
            try :
                date_1 =i.find('span','sclinkage_evalue').find_all('span')
                date_2 = date_1[1].get_text().strip()
            except :
                date_2='공개일가 없습니다'
                date2.append(date_2)
                print("4.공개일 : %s" %date_2)
            else :
                date2.append(date_2)
                print("4.공개일 : %s" %date_2)
            
            try :
                cont=soup_1.find("span","rp_indata").find('div','restd_incont').get_text().replace("\n","").strip()
            except :
                cont='상담내용이 없습니다'
                contents2.append(cont)
                print("5.상담내용 : %s" %cont)
            else :
                contents2.append(cont)
                print("5.상담내용 : %s" %cont)



            driver.back() # 이전 페이지로 돌아가기

            time.sleep(2)

            no += 1

            if no > cnt :
                break

    a += 1
    b = str(a)

    try :
        driver.find_element(By.LINK_TEXT ,'%s' %b).click()
    except :
        driver.find_element(By.LINK_TEXT,'다음 페이지로').click()

print("요청하신 작업이 모두 완료되었습니다")
```

```python

# Step 10. 수집된 데이터를 xlsx와 csv 형태로 저장하기
# 현재 날짜와 시간으로 폴더 만들고 파일 이름 설정하기
import os

n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' %(n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)

os.makedirs(f_dir+'RISS'+'-'+s+'-'+'학위논문')

fc_name = f_dir+'RISS'+'-'+s+'-'+'학위논문'+'\\'+'RISS'+'-'+s+'-'+'학위논문'+'.csv'
fx_name = f_dir+'RISS'+'-'+s+'-'+'학위논문'+'\\'+'RISS'+'-'+s+'-'+'학위논문'+'.xlsx'

# 데이터 프레임 생성 후 xlsx, csv 형식으로 저장하기
import pandas as pd

df = pd.DataFrame()
df['번호']=no2
df['제목']=pd.Series(title2)
df['날짜']=pd.Series(date2)
df['상담내용']=pd.Series(contents2)

# xls 형태로 저장하기
df.to_excel(fx_name,index=False)

# csv 형태로 저장하기
df.to_csv(fc_name,index=False, encoding="utf-8-sig")

print('요청하신 데이터 수집 작업이 정상적으로 완료되었습니다')
```
