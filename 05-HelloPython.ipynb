JSON(JavaScript Object Notation)은 경량의 데이터 교환 형식으로, 사람이 읽고 쓰기 쉬우며 기계가 분석하고 생성하기도 간편한 텍스트 기반 표준입니다. 자바스크립트 객체 문법을 기반으로 하지만 언어에 독립적이며, 이름-값 쌍(key-value pairs) 구조를 통해 웹 애플리케이션 등에서 서버와 데이터를 주고받을 때 주로 사용됩니다.

### 1. 딕셔너리의 이해 
- key와 value의 쌍으로 이루어진 자료구조
    - 변수이름 = { key1: value1, key2: value2, key3: value3 }
    - student = {"이름": "연주", "학년": 4, "전공": "경제학"}
    - 딕셔너리는 의미 중심 저장 :  student["이름"]
    - list[0]는 위치로 접근 / 딕셔너리는 dict["이름"] key로 접근 
        - 딕셔너리는 인덱스 개념이 없다. 
- key의 조건
    - 반드시 고유해야 한다 (중복 불가)
    - 변하지 않는 값이어야 한다
    - 보통 문자열 사용
    - value는 아무거나 가능
- 사용제약 
    - 존재하지 않는 값에 대한 출력이나 연산은 Error 
- 딕셔너리 확장/ 중복 제한 
    - 존재하지 않는 키에 값을 할당하면 확장 
    - 키가 중복될 경우 하나를 제외한 나머지는 무시됨 
        - 일반적으로 나중에 정의된 항목이 이전에 정의된 항목을 덮어 씀 
- 딕셔너리와 리스트는 필요에 따라 서로의 원소가 될 수 있다. 
    - 변수이름 = {"key" : [value1, value2, value3]} 
    - 변수이름["key"][index-number]

    - 딕셔너리1 = {"key1": value1, "key2":value2}
    - 딕셔너리2 = key3 : {"key1": value1, "key2":value2}
    - print(딕셔너리n["key1"]["key2"])  -> value2가 출력됨 
- '딕셔너리를 원소로 갖는 리스트' 를 원소로 포함하는 딕셔너리 ~ 상위 딕셔너리의 value가 됨 
    - {"key" " [ {  }, {  }, {  }...]}  ~ 모두 같은 구조로 존재해야 목록형태가 성립된다.
- 기본 데이터 타입과 연속성 자료형을 동시에 포함하는 딕셔너리

- 자주 겪는 Error

        - ZeroDivisionError : division by zero 
    - 리스트의 인덱스 접근 Error

- 고전적인 예외 처리

```python
# 잘못된 연산 에러 ( 어떤 수를 0으로 나누는 경우 )

# 입력값에 따라 에러가 발생하는 코드 

# x에 0이 입력된 경우 Error
    # ZeroDivisionError
    # ZeroDivisionError: division by zero
x = int(input("숫자를 입력하세요."))

y =100 /x 
print("결과는 %d입니다." % y)
```

```python
# 리스트의 인덱스 접근 Error 
# 반복문을 통해 리스트를 탐색 과정에서 인덱스의 범위를 초과하면 에러 
# IndexError: list index out of range

mylist = [92, 76, 83, 64, 95]
mysum = 0 

x = int(input("인원수를 입력하세요."))

# 0부터 입력값 x전까지 반복 
for i in range(x) : 
    mysum += mylist[i]

print("합계는 %d입니다." % mysum)
print("x = " , x)
```

```python
# 고전적인 예외 처리 (1)

# 연산을 수행할 수 없는 경우를 if문으로 대비해야 함 

# 정수만 입력 받아야 한다. 
x = int(input("숫자를 입력하세요."))

#
if x == 0:
    print("0으로 나눌 수 없습니다. ")   
else :
    y = 100/ x
    print("결과는 %d입니다." % y)
```

```python
# 고전적인 예외 처리 (2)

# 입력값이 범위를 초과하는지 미리 검사 해야 함 

mylist = [92, 76, 83, 64, 95]
mysum = 0 

x = int(input("인원수를 입력하세요."))

# 리스트의 길이는 1부터 세니까, 5
# 5보다 커야 하니까, 6부터 
if len(mylist) < x : 
    print("인원수가 너무 큽니다.")
else:
    #
    for i in range(x) :
        mysum += mylist[i]

    print("합계는 %d입니다." % mysum)
    print("x=", x)
```

```python
# try ~ except 문의 기본 활용 (1) 

# 예외상황이 예상되는 코드를 try블록으로 감싼다.
# 이 블록 안에서 예외가 발생할 경우, 연결된 except블록으로 프로그램 제어가 이동한다. 
# try블록 안에서 예외가 발생한 코드 이후의 라인은 실행되지 않는다. 
# 예외가 발생하더라도 프로그램이 비정상적으로 종료되지 않는다. 

try : 
    x = int(input("숫자를 입력하세요."))
    y = 100/ x 
    print("결과는 %d 입니다." % y)
except: 
    print("에러가 발생했습니다. ")
```

```python
# try ~ except 문의 기본 활용 (2) 

# 문자 입력 시
# 숫자를 입력하세요: abc
# ValueError 발생: invalid literal for int() with base 10: 'abc'

# 0 입력 시
# 숫자를 입력하세요: 0
# ZeroDivisionError 발생: division by zero

try : 
    x = int(input("숫자를 입력하세요."))
    y = 100/ x
    print("결과는 %d입니다." % y)
except ValueError: 
    print("숫자만 입력하세요.")
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except Exception: 
    print("알 수 없는 에러가 발생했습니다.")


```

```python
# 에러 메시지를 변수에 저장하기 (2)
# 숫자만 입력하세요. invalid literal for int() with base 10: 'abc'

try : 
    x = int(input("숫자를 입력하세요."))
    y = 100/ x
    print("결과는 %d입니다." % y)
except ValueError as verr:
    print("숫자만 입력하세요.", verr)
except ZeroDivisionError as zerr:
    print("0으로 나눌 수 없습니다.", zerr)
except Exception as err: 
    print("알 수 없는 에러가 발생했습니다.", err)
```

```python
#

# 숫자만 입력하세요. invalid literal for int() with base 10: 'abd'
# 처리가 완료되었습니다.

try : 
    x = int(input("숫자를 입력하세요."))
    y = 100/ x
    print("결과는 %d입니다." % y)
except ValueError as verr:
    print("숫자만 입력하세요.", verr)
except ZeroDivisionError as zerr:
    print("0으로 나눌 수 없습니다.", zerr)
except Exception as err: 
    print("알 수 없는 에러가 발생했습니다.", err)
finally:                    # 마지막에 실행될 코드 
    print("처리가 완료되었습니다.")
```

```python
# 가장 간편한 실행 형태 

# 예외 상황을 통합해서 대응할 수 있는 Exception클래스만 사용 

try : 
    x = int(input("숫자를 입력하세요."))
    y = 100/ x
    print("결과는 %d입니다." % y)
except Exception as e: 
    print("알 수 없는 에러가 발생했습니다.", e)
finally: 
    print("처리가 완료되었습니다.")
```

```python
#

# 알 수 없는 에러가 발생했습니다. 
# invalid literal for int() with base 10: 'fshdjksdhjk' 처리가 완료되었습니다.

mylist = [92, 76, 83, 64, 95]
mysum = 0 

try: 
    x = int(input("인원수를 입력하세요."))

    #
    for i in range(x):
        mysum += mylist[i]

    print("합계는 %d입니다." % mysum)
except Exception as e: 
    print("알 수 없는 에러가 발생했습니다.", e)
finally: 
    print("처리가 완료되었습니다.")
```

```python
# 이전 예제에서 생성한 파일을 읽어들이기 위해 연다. 
# -> 읽기 모드(r) 지정에 유의 
# -> 인코딩 방식은 저장 방식과 동일하게 지정. (다를 경우 에러)

# 파일의 모든 내용을 한 번에 문자열로 읽어들인다. 
# -> 아래와 같이 파일의 모든 내용을 변수에 저장한 것과 동일한 효과 
#          data = "Hello Python !!!\n"
f = open("helloworld.txt", "r", encoding='utf-8')

data =f.read()
print(data)

# open()함수를 단독으로 사용할 경우 파일작업 완료 후 닫기를 반드시 수행 
f.close()
```

```python
# 이전 예제와 동일하지만, f.close() 처리는 자동으로 수행함 
with open ("hellopython.txt", "w", encoding='utf-8') as f:
    # with 블록 안에서만 파일 객체 f가 유효함 
    for i in range(0,10): 
        f.write("%d>>" % i)
        f.write("Life is too short, ")
        f.write("you need python\n")
    
    print("파일 저장이 완료 되었습니다.")
```

```python
# 읽기 모드로 파일 객체 생성하기
# 'w'에서 'r'로 모드가 변경되어야 하므로 파일 객체를 새로 생성해야 한다. 
with open("hellopython.txt", "r", encoding='utf-8') as f:
    # 파일의 내용을 변수에 저장함 
    data = f.read()
    print(data)
```

```python
with open("helloworld.txt", "r", encoding='utf-8') as f:
    # 파일의 각 행을 원소로 갖는 리스트 생성
    # -> 아래의 구문과 동일한 효과 
    #       lines = ["Hello Python !!! \n", "안녕하세요. 파이썬!!!\n"]
    lines = f.readlines() 
    print(lines)
    print("-" * 30) 

    # 전체 리스트의 크기 확인 
    size = len(lines)
    print("읽어들인 데이터는 총 %d줄 입니다." % size)
    print("-" * 30) 

    for item in lines :
        # 읽어들인 데이터에 줄바꿈 문자가 포함되므로 공백제거 필요 
        print( item.strip ) 
```

# 5. CSV 파일 다루기
- csv 
    - 엑셀에서 사용하는 표 형식을 텍스트 형태로 변환한 파일 
    - 딕셔너리를 포함하는 리스트 구조를 CSV로 저장하기 


```python
# 한 학급의 성적표 데이터 예시 
# [리스트]
#   ↓
# [딕셔너리]
#   ↓
# [문자열 변환]
#   ↓
# [파일 기록]


# 1. 데이터 생성
# 2. 데이터 구조화
# 3. 저장 형식 선택
# 4. 직렬화 :  “직렬” = 일렬로 나열한다
    # 메모리에 있는 복잡한 객체를 저장 가능한 “일렬의 데이터”로 바꾸는 과정
    # data = {"name": "연주", "score": 95}
    # 이건 메모리 안에서만 이해되는 구조입니다. 이걸 파일에 저장하려면, 
    # 같은 문자열이나 바이트 형태로 바꿔야 합니다.
    # {"name": "연주", "score": 95}
# 5. 디스크 기록 : 직렬화된 데이터를 실제 저장장치에 쓰는 것
    # f.write("철수,95,88,72")
    # 이 순간: 운영체제에 저장 요청, 파일 시스템이 디스크에 기록, 물리적인 저장 완료

# 사람들이 말하는 저장은 보통: 직렬화 + 디스크 기록


#[메모리 객체]
#        ↓
#[직렬화]
#        ↓
#[디스크 기록]



grade = [
    {"name": "철수", "kor":95, "eng":88, "math":72},
    {"name": "영희", "kor":92, "eng":90, "math":95},
    {"name": "철민", "kor":88, "eng":76, "math":64}
]

# csv파일의 한 행에 대한 구문 형식 정의 
tpl = "{0},{1},{2},{3}\n"

# csv 파일 저장을 위한 f객체 생성 
# -> Excel의 csv는 euc-kr 형식 
# open(파일이름, 모드, 인코딩)
# Excel에서 한글 CSV 파일을 열 때

# with 문이 왜 중요하냐? 이 구조는: 자동으로 파일을 닫아줌, 메모리 누수 방지, 파일 안전하게 저장
# encoding 이란, 문자를 숫자로 바꿀 때 변환 규칙 
# EUC-KR = Extended Unix Code for Korean
# UTF = Unicode Transformation Format 
        # 전 세계 모든 문자를 저장할 수 있는 국제 표준 방식: 영어, 한글, 일본어, 중국어 모두 저장 가능
with open ("grade.csv", "w", encoding='euc-kr') as f:
    # 첫 줄에 각 항목의 제목 기록 
    # 저장시 파일에서 형태 : 이름, 국어, 영어, 수학 
    # 저장시 csv에서 형태  : 값,값,값  
    f.write("이름, 국어, 영어, 수학\n") 

    # 각 데이터를 한 줄씩 콤마로 구분하여 기록 
    # 여기서 item은 각 딕셔너리 하나씩을 의미합니다.
    for item in grade : 
        tmp = tpl.format(item ["name"], item["kor"], item["eng"], item["math"])
        f.write(tmp)










#--------------------------------------------------------------------
grade = [
    {"name": "철수", "kor":95, "eng":88, "math":72},
    {"name": "영희", "kor":92, "eng":90, "math":95},
    {"name": "철민", "kor":88, "eng":76, "math":64}
]

with open ("grade.csv", "w", encoding='euc-kr') as f:
    f.write("이름, 국어, 영어, 수학\n") 

    for item in grade : 
        tmp = tpl.format(item ["name"], item["kor"], item["eng"], item["math"])
        f.write(tmp)
```

```python
# 저장을 위한 자료구조 정의 
grade_dic = { 
    '이름' : ['철수', '영희', '민철', '수현', '호영'],
    '국어' : [98, 88, 92, 63, 120],
    '영어' : [76, 90, 70, 60, 50 ],
    '수학' : [88, 62, 88, 31, 76 ],
    '과학' : [64, 72, 45, 70, 88 ],
}

tpl = "{0}, {1}, {2}, {3}, {4}\n" # 한 줄에 대한 템플릿
keys = list(grade_dic.keys())     # 딕셔너리의 키를 리스트로 변환

p = ","
title = p.join(keys)    # keys리스트 사이에 쉼표를 넣어 문자열로 변환 


with open ("grade2.csv", "w", encoding='euc-kr') as f: 
    f.write(title + "\n") # 첫 줄에 각 항목의 제목 기록 

    # 각 데이터를 한 줄씩 콤마로 구분하여 기록
    # 들여쓰기를 유의해야 한다. 
    for i in range(0, len(grade_dic['이름'])):
        tmp = tpl.format(grade_dic['이름'][i], grade_dic['국어'][i], grade_dic['영어'][i], 
                     grade_dic['수학'][i], grade_dic['과학'][i] )
        f.write( tmp )
   




```

```python
#-------------------------------------
# step1. 파일 읽기 
#-------------------------------------
# 읽어들인 내용을 저장하기 위한 빈 변수 생성 
csv_list = None



# grade.csv 파일에 대한 파일 객체를 읽기 모드로 생성 
    # 파일의 각 행을 원소로 갖는 리스트 생성
    # -> ['이름,국어,영어,수학\n',
    #     '철수,95,88,72\n', 
    #     '영희,92,90,95\n',
    #     '철민,88,76,64\n']
with open ("grade.csv", "r", encoding='euc-kr') as f :
    csv_list = f.readlines() 

print(csv_list)

#-------------------------------------
# step2. 읽어온 내용을 변환하기  
#-------------------------------------
data = [] # csv의 데이터를 저장할 빈 배열 

for i, line in enumerate(csv_list) :
    # 첫 행은 제목이므로 건너 뛴다.
    if i ==0 : 
        continue 

    # 현재 행의 내용을 콤마를 기준으로 잘라서 리스트로 변환 
    # -> split()의 결과로 생성된 리스트의 모든 원소는 string 타입 
    # "철민,88,76, 64".split(",") ==> ["철민","88","76","64"]
    k = line.strip().split(",")

    # 한 행의 내용을 딕셔너리로 변환 
    # -> 과목별 점수는 정수로 형변환 후 처리 
    item = {"name":k[0], "kor":int(k[1]), "eng":int(k[2]), "math":int(k[3])}

    # 딕셔너리를 미리 준비한 리스트에 추가 
    data.append(item)

# 결과 확인 
data

```

```python
# covid19로 인한 서울시와 전국 단위 일평균 확진/사망자 수(1)
#-----------------------------------------------------------
# step 1 
#-----------------------------------------------------------
# 예외처리와 with문을 사용한 코드 영역 안에서 csv 파일 읽어오기 
# csv 파일의 구조 특성상 각 행을 리스트로 변환해야 한다. 


# try ~ with / except / finally 
# 파일을 열자 / csv 파일이니까, 인코딩 euc-kr 
# 잘 열리는지 몇 개 열어보기 ~ 반복문 for 
# 
try : 
    with open("covid19.csv", 'r', encoding='euc-kr') as f : 
        # csv 파일의 각 행을 원소로 갖는 리스트 생성 
        csv_list = f.readlines()
        # step 1 
        # 상위 5줄 확인 
        for i in range(5) :                     # 일단 잘 불러와 지는지, 5줄만 확인해보자 
            print(csv_list[i].strip())

except Exception as e:
    print("CSV 파일 처리에 실패했습니다.", e)
finally : 
    print("-" * 30) 
    print("작업이 완료되었습니다.")
```

```python
# covid19로 인한 서울시와 전국 단위 일평균 확진/사망자 수(2)
#-----------------------------------------------------------
# step 2
#-----------------------------------------------------------

try : 
    with open("covid19.csv", 'r', encoding='euc-kr') as f : 
        # csv 파일의 각 행을 원소로 갖는 리스트 생성 
        csv_list = f.readlines()
        # step 1 
        # 상위 5줄 확인 
        for i in range(5) :                     # 일단 잘 불러와 지는지, 5줄만 확인해보자 

        # 전체 데이터 길이 (제목행을 제외하기 위해 -1 처리) 
        size = len(csv_list) -1 

        # step 2 
        # 합계를 위한 변수 초기화 ~ 누적합을 구하기 위한 준비 단계 
        seoul_confirm_sum = 0       # 서울 확진자 수 합계 
        seoul_death_sum = 0         # 서울 사망자 수 합계 
        nation_confirm_sum = 0      # 전국 확진자 수 합계 
        nation_death_sum = 0        # 전국 사망자 수 합계 


except Exception as e:
    print("CSV 파일 처리에 실패했습니다.", e)
finally : 
    print("-" * 30) 
    print("작업이 완료되었습니다.")
```

```python
# covid19로 인한 서울시와 전국 단위 일평균 확진/사망자 수(3)
#-----------------------------------------------------------
# step 3
#-----------------------------------------------------------



try : 
    with open("covid19.csv", 'r', encoding='euc-kr') as f : 
        # csv 파일의 각 행을 원소로 갖는 리스트 생성 
        csv_list = f.readlines()
        # step 1 
        # 상위 5줄 확인 
        #for i in range(5) :                     # 일단 잘 불러와 지는지, 5줄만 확인해보자 

        # 전체 데이터 길이 (제목행을 제외하기 위해 -1 처리) 
        size = len(csv_list) -1 

        # step 2 
        # 합계를 위한 변수 초기화 
        seoul_confirm_sum = 0       # 서울 확진자 수 합계 
        seoul_death_sum = 0         # 서울 사망자 수 합계 
        nation_confirm_sum = 0      # 전국 확진자 수 합계 
        nation_death_sum = 0        # 전국 사망자 수 합계 

        # step 3
        # csv 파일의 각 행을 순회하며 데이터 추출 
        for i, line in enumerate(csv_list) :
            # 첫 행은 제목이므로 건너 뛴다. 
            if i == 0 : 
                continue 
    
            # 현재 행의 내용을 콤마를 기준으로 잘라서 리스트로 변환 
            k = line.strip().split(",")
        
            # 서울의 확진자 수와 사망자 수 추출 
            if k[1].isnumeric() : seoul_confirm_sum += int(k[1])
            if k[2].isnumeric() : seoul_death_sum += int(k[2])
        
            # 전국의 확진자 수와 사망자 수 추출 
            if k[3].isnumeric() : nation_confirm_sum += int(k[3])
            if k[4].isnumeric() : nation_death_sum += int(k[4])
        

        # 일 평균 출력 
        print("서울 일 평균 확진자 수 : %.2f명" % (seoul_confirm_sum / size))
        print("서울 일 평균 사망자 수 : %.2f명" % (seoul_death_sum / size))
        print("서울 일 평균 확진자 수 : %.2f명" % (seoul_confirm_sum / size))  
        print("서울 일 평균 사망자 수 : %.2f명" % (seoul_death_sum / size))    
              



except Exception as e:
    print("CSV 파일 처리에 실패했습니다.", e)
finally : 
    print("-" * 30) 
    print("작업이 완료되었습니다.")
```
