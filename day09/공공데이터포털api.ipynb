```python
# 필요한 라이브러리 import 하기
import requests
import json

# Open API Key를 포함한 url 설정
url = 'https://api.odcloud.kr/api/15070282/v1/uddi:00e5cb5a-ecdf-4190-a499-ba3a6b2a8db9?page=1&perPage=10&serviceKey=03f9e63dd117122e2042d5dd3789ce368597ded05d6b1976baf7a3fd6c182de9'

# url에 접속 후 데이터 불러오기
response = requests.get(url)

# 불러온 데이터 확인하기
contents = response.text
print(contents)

```

```python
# 불러온 데이터를 json 형태로 변경
json_data = json.loads(contents)
print(json_data)
print()

print('데이터타입:', type(json_data)) #json타입 확인

```

```python
# 필요한 값만 추출하기
data = json_data['data']

# Dataframe으로 만들기
import pandas as pd
df = pd.json_normalize(data)

df
```
