```python
# í”„ë¡œí¼í‹°ë¥¼ ì •ì˜í•œ í´ë˜ìŠ¤ 

# ë©¤ë²„ë³€ìˆ˜ì™€ getter, setter ì •ì˜í•˜ê¸° 
# ë©¤ë²„ë³€ìˆ˜ __name ê³¼ __hpë¥¼ ì •ì˜í•˜ê³  
# __nameì— ëŒ€í•œ getter , setterë¥¼ ì¶”ê°€í•œë‹¤. 

class Vulture:
    __name = None
    __hp = None 

    # __nameì— ëŒ€í•œ í”„ë¡œí¼í‹° ì •ì˜ 
    @property 
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, value):
        self.__name = value
    
    @property 
    def hp(self) :
        return self.__hp 

    @hp.setter 
    def hp(self, value):
        if value < 1 :
            value = 1
        self.__hp = value
    
    #ê°ì²´ì˜ íŠ¹ì„±ì„ ì´ˆê¸°í™” í•˜ê¸° ìœ„í•œ ìƒì„±ì ì •ì˜ 
    def __init__(self, name = None, hp =0, dps=0):
        #setter í”„ë¡œí¼í‹°ì—ê²Œ íŒŒë¼ë¯¸í„°ë¥¼ ì „ë‹¬í•˜ì—¬ ë©¤ë²„ë³€ìˆ˜ ì´ˆê¸°í™” 
        self.name = name 
        self.hp = hp 
        print("[{0}] Alright! Bring it on!".format(self.name))

    # ê°ì²´ì˜ ì†ì„±ì„ ì¼ê´„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ 
    def state(self): 
        #getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ì¶œë ¥í•œë‹¤. 
        print("[{0}] hp : {1}".format(self.name, self.hp))

    # ë ˆë²¨ì—… --> ëª¨ë“  ì†ì„±(ë©¤ë²„ë³€ìˆ˜)ë“¤ì´ ì¦ê°€í•œë‹¤. 
    def level_up(self):
        #getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ì¦ê°€ì‹œí‚¨ ê²°ê³¼ë¥¼ setterë¥¼ í†µí•´ ì €ì¥í•œë‹¤. 
        self.hp += 5
        print("[{0}] ë ˆë²¨ì—…ì„ í•˜ì—¬ ì²´ë ¥(+5)ì´ ì¦ê°€í•©ë‹ˆë‹¤.".format(self.name))

    # ê³µê²©ì„ ë°›ìŒ --> ìƒëŒ€ì˜ ê³µê²©ë ¥ë§Œí¼ hpê°€ ê°ì†Œí•œë‹¤. 
    def attacked(self, enemy_attack):
        # getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ê°ì†Œì‹œí‚¨ ê²°ê³¼ë¥¼ setterë¥¼ í†µí•´ ì €ì¥í•œë‹¤. 
        self.hp -= enemy_attack
        print("[{0}]ì ì˜ ê³µê²©ì„ ë°›ì•„ hpê°€ {1}ê°ì†Œí•©ë‹ˆë‹¤.".format(self.name, enemy_attack))    
```

```python
# ìƒì„±ìì— ì •ì˜ëœ dpsì˜ ê¸°ë³¸ê°’ì´ ì‘ìš©í•˜ì—¬ dpsê°€ 0ìœ¼ë¡œ ì„¤ì •ëœë‹¤. 
# dpsê°€ setterì— ì „ë‹¬ë˜ì–´ ifë¬¸ì— ì˜í•´ 1ë¡œ ì¡°ì •ëœë‹¤. 

v1 = Vulture("ë²Œì²˜1" , 100)
v2 = Vulture("ë²Œì²˜2" , 100, 20)

v1.state()
v2.state()

v1.attacked(30)
v2.attacked(20)

v1.state()
v2.state()

v1.level_up()
v2.level_up()


v1.state()
v2.state()
```

```python
# í´ë˜ì‹œ ìƒì†ì˜ ì´í•´ 

# class Calc(ë¶€ëª¨í´ë˜ìŠ¤ì´ë¦„): 
# í´ë˜ìŠ¤ Aë¥¼ í´ë˜ìŠ¤ Bê°€ ìƒì† ë°›ì„ë•Œ, 
# BëŠ” Aì˜ ëª¨ë“  ë©¤ë²„ë³€ìˆ˜ì™€ ë©”ì„œë“œë¥¼ ìì‹ ì˜ ê²ƒìœ¼ë¡œ ìƒì†ë°›ê²Œ ëœë‹¤. 


# ìƒì†ì„ í†µí•œ í´ë˜ìŠ¤ ê¸°ëŠ¥ í™•ì¥ 
# ë¶€ëª¨ í´ë˜ìŠ¤ ì •ì˜í•˜ê¸° 
class Terran : 
    name = None     # ìœ ë‹›ì˜ ì´ë¦„ 
    hp = None       # ìœ ë‹›ì˜ ì²´ë ¥ 

    def __init__(self, name, hp): 
        self.name = name 
        self.hp = hp 

    def info(self):
        return "[%s] HP : %d" % (self.name, self.hp)        #ë¬¸ìì—´(String) ì •ìˆ˜(decimal integer)
```

```python
# Marineì€ Terran ì„ ìƒì†ë°›ì•„ ê³µê²© ê¸°ëŠ¥ì„ ì¶”ê°€í•œë‹¤. 

class Marine(Terran) :          #Marineì€ Terranì˜ ëª¨ë“  ê¸°ëŠ¥ì„ ë¬¼ë ¤ë°›ëŠ”ë‹¤.
    # ê¸°ëŠ¥ì˜ í™•ì¥ì„ ìœ„í•´ ì¶”ê°€ëœ ë©”ì„œë“œ 
    def shoot (self, attack ) :
        return "[%s] ì´ì„ ì˜ì•„ %dì˜ ë°ë¯¸ì§€ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤." %(self.name, attack) 


# ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ìƒì„±ìë¥¼ ìƒì†ë°›ê¸° ë•Œë¬¸ì— nameê³¼ hp íŒŒë¼ë¯¸í„°ë¥¼ ì „ë‹¬í•  ìˆ˜ ìˆë‹¤. 
m = Marine('ë§ˆë¦°1', 45)     # Marine ì•ˆì—ëŠ” __init__()ì´ ì •ì˜ë˜ì–´ ìˆì§€ ì•ŠìŒ, Terran.__init__(m, 'ë§ˆë¦°1', 45)
print(m.info())         # Marineì—ëŠ” infoê°€ ì—†ìœ¼ë¯€ë¡œ Terranì˜ info() ì‚¬ìš©
print(m.shoot(100))
```

```python
# super()ë¥¼ ì‚¬ìš©í•œ ë¶€ëª¨ê¸°ëŠ¥ í˜¸ì¶œí•˜ê¸° 

class Medic(Terran) : 
    mana = None     # **ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ë©¤ë²„ë³€ìˆ˜ ì´ì™¸ì— ì¶”ê°€ ì†ì„±** Medicì€ ë¶€ëª¨ë³´ë‹¤ ì†ì„±ì´ í•˜ë‚˜ ë” ë§ë‹¤

    def __init__(self, name, hp, mana) : 
        # ìŠ¤ìŠ¤ë¡œì˜ ìƒì„±ìì— ì „ë‹¬ë°›ì€ íŒŒë¼ë¯¸í„°ë¥¼ ë¶€ëª¨ ìƒì„±ìì—ê²Œ ì „ë‹¬í•˜ì—¬ 
        # ë¶€ëª¨ë¡œ ë¶€í„° ìƒì†ë°›ì€ ì†ì„±(name, hp)ë¥¼ ì´ˆê¸°í™” í•œë‹¤. 
        super().__init__(name, hp)      #ë¶€ëª¨ ìƒì„±ìë¥¼ ì§ì ‘ í˜¸ì¶œí•˜ëŠ” ë°©ë²•ì´ super()

        # ìŠ¤ìŠ¤ë¡œ ì¶”ê°€í•œ ì†ì„±ì„ ë³„ë„ë¡œ ì´ˆê¸°í™” í•œë‹¤. 
        self.mana = mana 

    # ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ê¸°ëŠ¥ì„ í¬í•¨í•˜ëŠ” ì¶”ê°€ ê¸°ëŠ¥ 
    def info_ex(self):
        info_msg = super().info()       #ë¶€ëª¨ í´ë˜ìŠ¤ì˜ info() í˜¸ì¶œ
        info_msg += ", MANA: %d" % self.mana    # [ë©”ë”•1] HP : 30"ì„ ë¨¼ì € ë§Œë“  ë‹¤ìŒ ", MANA: 100"ì„ ë§ë¶™ì´ëŠ” êµ¬ì¡°ì•¼.
        return info_msg 

    # ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ê¸°ëŠ¥ ì´ì™¸ì— ìì‹ ë§Œì˜ ê³ ìœ  ê¸°ëŠ¥ ì¶”ê°€ 
    def heel(self, target) :
        self.mana -= 5
        return "[%s] %sì˜ ì²´ë ¥ì„ íšŒë³µì‹œí‚µë‹ˆë‹¤. (mana: -5)" % (self.name, target)

md = Medic('ë©”ë”•1', 30, 100) 
print(md.info_ex())
print(md.heel('ë§ˆë¦°1'))
print(md.info_ex())
```

```python
# ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë“œ overrideì˜ ì´í•´ 

# ë¶€ëª¨í´ë˜ìŠ¤ ê¸°ëŠ¥ ìˆ˜ì •í•˜ê¸° 
# ìì‹ í´ë˜ìŠ¤ê°€ ë¶€ëª¨ì™€ ë™ì¼í•œ ì´ë¦„ì˜ ë©”ì„œë“œë¥¼ ì •ì˜í•˜ëŠ” í˜•íƒœ 
# ìƒì†ë°›ì€ ë©”ì„œë“œë¥¼ 'ì¬ì •ì˜ override'í•˜ì—¬ ìì‹ í´ë˜ìŠ¤ì˜ í–‰ë™ì„ ë¶€ëª¨ì™€ ë‹¤ë¥´ê²Œ ë°”ê¿€ ìˆ˜ ìˆë‹¤. 



# ë¶€ëª¨í´ë˜ìŠ¤ì˜ ê¸°ëŠ¥ì„ ìœ ì§€í•˜ë©´ì„œ ì¶”ê°€ì ì¸ ê¸°ëŠ¥ ë§ë¶™ì´ê¸° 
# ë¶€ëª¨ì˜ ë©”ì„œë“œë¥¼ override í•˜ëŠ” ë©”ì„œë“œ ì•ˆì—ì„œëŠ” 
# super() í•¨ìˆ˜ë¥¼ í†µí•´ ì¬ì •ì˜ ë˜ê¸° ì „ì˜ ì›ë³¸ ê¸°ëŠ¥ì— ì ‘ê·¼í•  ìˆ˜ ìˆë‹¤. 
class Animal : 
    def sound(self) :
        print("ë™ë¬¼ì´ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤.")      # í•¨ìˆ˜ëŠ” ì •ì˜ë˜ì—ˆë‹¤ê³  í•´ì„œ ì‹¤í–‰ë˜ëŠ” ê²ƒì´ ì•„ë‹˜. "í˜¸ì¶œë  ë•Œë§Œ" ì‹¤í–‰ë¨

class Cat(Animal) : 
    def sound(self) :
        super().sound()     # ë¶€ëª¨ ê¸°ëŠ¥ í˜¸ì¶œ, ë¶€ëª¨ í´ë˜ìŠ¤(Animal)ì˜ sound()ë¥¼ ì‹¤í–‰í•˜ë¼, ë™ë¬¼ì´ ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤. í•œë²ˆ ì¶œë ¥ë¨. 
        print("ì•¼ì˜¹~") 

c = Cat()       # Cat ê°ì²´ ìƒì„± , Catì€ Animalì„ ìƒì†ë°›ì•˜ìœ¼ë¯€ë¡œ ë‚´ë¶€ì ìœ¼ë¡œ Animal êµ¬ì¡°ë„ í¬í•¨
c.sound()       # Cat í´ë˜ìŠ¤ì—ì„œ sound ì°¾ìŒ â†’ ìˆìŒ, Catì˜ sound() ì‹¤í–‰
Animal.sound(c)
```

```python
class Terran : 
    name = None     # ìœ ë‹›ì˜ ì´ë¦„ 
    hp = None       # ìœ ë‹›ì˜ ì²´ë ¥ 

    def __init__(self, name, hp): 
        self.name = name 
        self.hp = hp 

    def info(self):
        return "[%s] HP : %d" % (self.name, self.hp)        #ë¬¸ìì—´(String) ì •ìˆ˜(decimal integer)
```

```python
# í•˜ìœ„ í´ë˜ìŠ¤ë¥¼ í†µí•œ ë¶€ëª¨ ê¸°ëŠ¥ì˜ ìˆ˜ì • 
# SiegeTank ì€ Terranì„ ìƒì†í•˜ê³  , ì¶”ê°€ ìƒíƒœ(ì‹œì¦ˆëª¨ë“œ)ë¥¼ ê´€ë¦¬í•œë‹¤. 

class SiegeTank(Terran) :
    siege_mode = None

    def __init__(self, name, hp=150, siege_mode=False) :
        super().__init__(name, hp)      # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ìƒì„±ìë¥¼ ì‹¤í–‰í•˜ë¼, self.name = name, self.hp = hp
        self.siege_mode = siege_mode    # ìì‹ í´ë˜ìŠ¤ë§Œì˜ ìƒˆë¡œìš´ ì†ì„± ì¶”ê°€

    def set_siege(self, on):            # on ê°’ì„ True/Falseë¡œ ë³€í™˜, ìƒíƒœ ë³€ê²½ ë©”ì„œë“œ
        self.siege_mode = bool(on)

    # Override :: ë¶€ëª¨ í´ë˜ìŠ¤ê°€ ê°–ê³  ìˆëŠ” ë©”ì„œë“œë¥¼ ì¬ì •ì˜ í•œë‹¤. --> ê¸°ëŠ¥ì˜ ìˆ˜ì • 
    def info(self): 
        base = super().info()   # super().info() ì‹¤í–‰, ë¶€ëª¨ì˜ info() ì‹¤í–‰, "[íƒ±í¬1] HP : 150" ë°˜í™˜, baseì— ì €ì¥

        if self.siege_mode:
            mode = 'ì‹œì¦ˆëª¨ë“œ'
        else: 
            mode = 'ì•„ë¨¸ë“œëª¨ë“œ'
        
        return base + ", ìƒíƒœ: " + mode 
    

t = SiegeTank('íƒ±í¬1')  
print(t.info())
t.set_siege(True)
print(t.info())

```
