```python
# ë©”ì„œë“œë¥¼ í†µí•œ ë©¤ë²„ë³€ìˆ˜ì˜ ë°ì´í„° ì„¤ì • ë° ì¶œë ¥ 
# ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ìˆ˜ì •í•˜ê¸° ìœ„í•œ ë©”ì„œë“œë¥¼ ê°–ëŠ” í´ë˜ìŠ¤ 

# "Zealotì´ë¼ëŠ” ì„¤ê³„ë„ë¥¼ ë§Œë“¤ê³ ,
# ê·¸ ì„¤ê³„ë„ë¡œ zealot1ì´ë¼ëŠ” ê°ì²´ë¥¼ ë§Œë“  ë‹¤ìŒ,
# ê·¸ ê°ì²´ì˜ ìƒíƒœë¥¼ ë°”ê¾¸ëŠ” í–‰ë™ì„ ë©”ì„œë“œë¡œ ì‹¤í–‰í•œë‹¤."
class Zealot:           # "Zealot ê°ì²´ëŠ” ì²´ë ¥(hp)ê³¼ ê³µê²©ë ¥(dps)ì´ë¼ëŠ” ì†ì„±ì„ ê°€ì§„ë‹¤."
    hp = None   # ì²´ë ¥
    dps = None  # ê³µê²©ë ¥

    # ê°ì²´ì˜ ì†ì„±ì„ ì¼ê´„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ 
    def state(self):
        print("ìƒíƒœí™•ì¸ >>> hp: {0}, dps: {1}.format(self.hp, self.dps)")


    # ë ˆë²¨ì—… --> ëª¨ë“  ì†ì„±(ë©¤ë²„ë³€ìˆ˜)ë“¤ì´ ì¦ê°€í•œë‹¤. 
    def level_up(self):
        self.hp += 10 
        self.dps += 2
        print("ë ˆë²¨ì—…ì„ í•˜ì—¬ ì²´ë ¥(+10), ê³µê²©ë ¥(+2)ì´ ì¦ê°€í•©ë‹ˆë‹¤.")



    # ê³µê²©ì„ ë°›ìŒ --> ìƒëŒ€ì˜ ê³µê²©ë ¥ ë§Œí¼ hpê°€ ê°ì†Œí•œë‹¤. 
    # enemy_attackì€ ì™¸ë¶€ì—ì„œ ì „ë‹¬ë°›ëŠ” ê°’
    def attacked(self, enemy_attack):
        self.hp -= enemy_attack
        print("{0}ì˜ ê³µê²©ì„ ë°›ì•„ hpê°€ {1}ë§Œí¼ ê°ì†Œí•©ë‹ˆë‹¤.".format(enemy_attack, self.hp))

```

```python
# ìƒì„±ìë¥¼ ê°–ëŠ” í´ë˜ìŠ¤ ì •ì˜í•˜ê¸° 

# í´ë˜ìŠ¤ ì •ì˜ (ì„¤ê³„ë„ ê·¸ë¦¬ê¸°)
# ê°ì²´ ìƒì„± (ì‹¤ì œ ì¸ìŠ¤í„´ìŠ¤ ë§Œë“¤ê¸°)
# ë©”ì„œë“œ í˜¸ì¶œ (í–‰ë™ ìˆ˜í–‰)

# ë§ˆë¦° í´ë˜ìŠ¤ ì •ì˜ 
class Marine: 
    name = None
    hp = None

    # íŒŒë¼ë¯¸í„°ë¥¼ í™œìš©í•´ ê°ì²´ì˜ ì†ì„±ì„ ì´ˆê¸°í™” í•˜ëŠ” ìƒì„±ì 
    def __init__(self, name, hp):
        self.name = name
        self.hp = hp 
        print("[{0} You wanna piece of me, boy?.format(self.name)]")    

    # ê°ì²´ì˜ ì†ì„±ì„ ì¼ê´„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ 
    def state(self):
        print("[{0}] hp: {1}".format(self.name, self.name))

    # ë ˆë²¨ì—… --> ëª¨ë“  ì†ì„±(ë©¤ë²„ë³€ìˆ˜)ë“¤ì´ ì¦ê°€í•œë‹¤. 
    def level_up(self):
        self.hp += 5
        print("[{0}] ë ˆë²¨ì—…ì„ í•˜ì—¬ ì²´ë ¥(+5)ì´ ì¦ê°€í•©ë‹ˆë‹¤.".format(self.name))

    # ê³µê²©ì„ ë°›ìŒ --> ìƒëŒ€ì˜ ê³µê²©ë ¥ ë§Œí¼ hpê°€ ê°ì†Œí•œë‹¤. 
    def attacked(self, enemy_attack):
        self.hp -= enemy_attack
        print("[{0}]ì ì˜ ê³µê²©ì„ ë°›ì•„ hpê°€ {1} ê°ì†Œí•©ë‹ˆë‹¤.".format(self.name, enemy_attack))

```

## 3. ì •ë³´ì€ë‹‰(ìº¡ìŠí™”)

```python
# í”„ë¡œí¼í‹°ë¥¼ ì •ì˜í•œ í´ë˜ìŠ¤ 

# ë©¤ë²„ë³€ìˆ˜ì™€ getter, setter ì •ì˜í•˜ê¸° 
# ë©¤ë²„ë³€ìˆ˜ __name ê³¼ __hpë¥¼ ì •ì˜í•˜ê³  
# __nameì— ëŒ€í•œ getter , setterë¥¼ ì¶”ê°€í•œë‹¤. 

# ê°ì²´ ë‚´ë¶€ ë°ì´í„°(__name, __hp)ë¥¼ ì§ì ‘ ê±´ë“œë¦¬ì§€ ëª»í•˜ê²Œ í•˜ê³ ,
# ë°˜ë“œì‹œ ì •í•´ì§„ í†µë¡œ(property)ë¥¼ í†µí•´ì„œë§Œ ìˆ˜ì •í•˜ê²Œ ë§Œë“ ë‹¤.

# 1. ë‚´ë¶€ì— ìˆ¨ê²¨ì§„ ë³€ìˆ˜(__name, __hp)ê°€ ìˆë‹¤.
# 2. ê·¸ ë³€ìˆ˜ì— ì ‘ê·¼í•˜ëŠ” ê³µì‹ ì°½êµ¬(property)ê°€ ìˆë‹¤.
# 3. ìƒì„±ìì—ì„œ ê·¸ ì°½êµ¬ë¥¼ í†µí•´ ì´ˆê¸°í™”í•œë‹¤.
# 4. ë©”ì„œë“œì—ì„œë„ ì§ì ‘ ë³€ìˆ˜ ì ‘ê·¼ì´ ì•„ë‹ˆë¼ propertyë¥¼ ì‚¬ìš©í•œë‹¤.

# [ì™¸ë¶€ ì½”ë“œ]
#      â†“
#   (name setter)
#      â†“
#   __name (ìˆ¨ê²¨ì§„ ì‹¤ì œ ë°ì´í„°)


class Vulture:
    __name = None
    __hp = None 

    # __nameì— ëŒ€í•œ í”„ë¡œí¼í‹° ì •ì˜ 
    @property 
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, value):
        self.__name = value
    
    @property 
    def hp(self) :
        return self.__hp 

    @hp.setter 
    def hp(self, value):
        if value < 1 :
            value = 1
        self.__hp = value
    
    #ê°ì²´ì˜ íŠ¹ì„±ì„ ì´ˆê¸°í™” í•˜ê¸° ìœ„í•œ ìƒì„±ì ì •ì˜ 
    def __init__(self, name = None, hp =0, dps=0):
        #setter í”„ë¡œí¼í‹°ì—ê²Œ íŒŒë¼ë¯¸í„°ë¥¼ ì „ë‹¬í•˜ì—¬ ë©¤ë²„ë³€ìˆ˜ ì´ˆê¸°í™” 
        self.name = name 
        self.hp = hp 
        print("[{0}] Alright! Bring it on!".format(self.name))

    # ê°ì²´ì˜ ì†ì„±ì„ ì¼ê´„ ì¶œë ¥í•˜ëŠ” ë©”ì„œë“œ 
    def state(self): 
        #getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ì¶œë ¥í•œë‹¤. 
        print("[{0}] hp : {1}".format(self.name, self.hp))

    # ë ˆë²¨ì—… --> ëª¨ë“  ì†ì„±(ë©¤ë²„ë³€ìˆ˜)ë“¤ì´ ì¦ê°€í•œë‹¤. 
    def level_up(self):
        #getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ì¦ê°€ì‹œí‚¨ ê²°ê³¼ë¥¼ setterë¥¼ í†µí•´ ì €ì¥í•œë‹¤. 
        self.hp += 5
        print("[{0}] ë ˆë²¨ì—…ì„ í•˜ì—¬ ì²´ë ¥(+5)ì´ ì¦ê°€í•©ë‹ˆë‹¤.".format(self.name))

    # ê³µê²©ì„ ë°›ìŒ --> ìƒëŒ€ì˜ ê³µê²©ë ¥ë§Œí¼ hpê°€ ê°ì†Œí•œë‹¤. 
    def attacked(self, enemy_attack):
        # getterë¥¼ í†µí•´ ë©¤ë²„ë³€ìˆ˜ì˜ ê°’ì„ ë°˜í™˜ë°›ì•„ ê°ì†Œì‹œí‚¨ ê²°ê³¼ë¥¼ setterë¥¼ í†µí•´ ì €ì¥í•œë‹¤. 
        self.hp -= enemy_attack
        print("[{0}]ì ì˜ ê³µê²©ì„ ë°›ì•„ hpê°€ {1}ê°ì†Œí•©ë‹ˆë‹¤.".format(self.name, enemy_attack))    
```

```python
# ìƒì„±ìì— ì •ì˜ëœ dpsì˜ ê¸°ë³¸ê°’ì´ ì‘ìš©í•˜ì—¬ dpsê°€ 0ìœ¼ë¡œ ì„¤ì •ëœë‹¤. 
# dpsê°€ setterì— ì „ë‹¬ë˜ì–´ ifë¬¸ì— ì˜í•´ 1ë¡œ ì¡°ì •ëœë‹¤. 

v1 = Vulture("ë²Œì²˜1" , 100)
v2 = Vulture("ë²Œì²˜2" , 100, 20)

v1.state()
v2.state()

v1.attacked(30)
v2.attacked(20)

v1.state()
v2.state()

v1.level_up()
v2.level_up()


v1.state()
v2.state()
```

```python
# super()ë¥¼ ì‚¬ìš©í•œ ë¶€ëª¨ê¸°ëŠ¥ í˜¸ì¶œí•˜ê¸° 

class Medic(Terran) : 
    mana = None     # **ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ë©¤ë²„ë³€ìˆ˜ ì´ì™¸ì— ì¶”ê°€ ì†ì„±** Medicì€ ë¶€ëª¨ë³´ë‹¤ ì†ì„±ì´ í•˜ë‚˜ ë” ë§ë‹¤

    def __init__(self, name, hp, mana) : 
        # ìŠ¤ìŠ¤ë¡œì˜ ìƒì„±ìì— ì „ë‹¬ë°›ì€ íŒŒë¼ë¯¸í„°ë¥¼ ë¶€ëª¨ ìƒì„±ìì—ê²Œ ì „ë‹¬í•˜ì—¬ 
        # ë¶€ëª¨ë¡œ ë¶€í„° ìƒì†ë°›ì€ ì†ì„±(name, hp)ë¥¼ ì´ˆê¸°í™” í•œë‹¤. 
        super().__init__(name, hp)      #ë¶€ëª¨ ìƒì„±ìë¥¼ ì§ì ‘ í˜¸ì¶œí•˜ëŠ” ë°©ë²•ì´ super()

        # ìŠ¤ìŠ¤ë¡œ ì¶”ê°€í•œ ì†ì„±ì„ ë³„ë„ë¡œ ì´ˆê¸°í™” í•œë‹¤. 
        self.mana = mana 

    # ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ê¸°ëŠ¥ì„ í¬í•¨í•˜ëŠ” ì¶”ê°€ ê¸°ëŠ¥ 
    def info_ex(self):
        info_msg = super().info()       #ë¶€ëª¨ í´ë˜ìŠ¤ì˜ info() í˜¸ì¶œ
        info_msg += ", MANA: %d" % self.mana    # [ë©”ë”•1] HP : 30"ì„ ë¨¼ì € ë§Œë“  ë‹¤ìŒ ", MANA: 100"ì„ ë§ë¶™ì´ëŠ” êµ¬ì¡°ì•¼.
        return info_msg 

    # ë¶€ëª¨ì—ê²Œ ìƒì†ë°›ì€ ê¸°ëŠ¥ ì´ì™¸ì— ìì‹ ë§Œì˜ ê³ ìœ  ê¸°ëŠ¥ ì¶”ê°€ 
    def heel(self, target) :
        self.mana -= 5
        return "[%s] %sì˜ ì²´ë ¥ì„ íšŒë³µì‹œí‚µë‹ˆë‹¤. (mana: -5)" % (self.name, target)

md = Medic('ë©”ë”•1', 30, 100) 
print(md.info_ex())
print(md.heel('ë§ˆë¦°1'))
print(md.info_ex())
```

```python
# í•˜ìœ„ í´ë˜ìŠ¤ë¥¼ í†µí•œ ë¶€ëª¨ ê¸°ëŠ¥ì˜ ìˆ˜ì • 
# SiegeTank ì€ Terranì„ ìƒì†í•˜ê³  , ì¶”ê°€ ìƒíƒœ(ì‹œì¦ˆëª¨ë“œ)ë¥¼ ê´€ë¦¬í•œë‹¤. 

class SiegeTank(Terran) :
    siege_mode = None

    def __init__(self, name, hp=150, siege_mode=False) :
        super().__init__(name, hp)      # ë¶€ëª¨ í´ë˜ìŠ¤ì˜ ìƒì„±ìë¥¼ ì‹¤í–‰í•˜ë¼, self.name = name, self.hp = hp
        self.siege_mode = siege_mode    # ìì‹ í´ë˜ìŠ¤ë§Œì˜ ìƒˆë¡œìš´ ì†ì„± ì¶”ê°€

    def set_siege(self, on):            # on ê°’ì„ True/Falseë¡œ ë³€í™˜, ìƒíƒœ ë³€ê²½ ë©”ì„œë“œ
        self.siege_mode = bool(on)

    # Override :: ë¶€ëª¨ í´ë˜ìŠ¤ê°€ ê°–ê³  ìˆëŠ” ë©”ì„œë“œë¥¼ ì¬ì •ì˜ í•œë‹¤. --> ê¸°ëŠ¥ì˜ ìˆ˜ì • 
    def info(self): 
        base = super().info()   # super().info() ì‹¤í–‰, ë¶€ëª¨ì˜ info() ì‹¤í–‰, "[íƒ±í¬1] HP : 150" ë°˜í™˜, baseì— ì €ì¥

        if self.siege_mode:
            mode = 'ì‹œì¦ˆëª¨ë“œ'
        else: 
            mode = 'ì•„ë¨¸ë“œëª¨ë“œ'
        
        return base + ", ìƒíƒœ: " + mode 
    

t = SiegeTank('íƒ±í¬1')  
print(t.info())
t.set_siege(True)
print(t.info())

```
