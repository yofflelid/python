```python
# 폴더 트리 생성 및 삭제 
import shutil # 파일, 폴더 관련 작업에 확장된 기능을 제공

# 현재 작업 위치에서 "python"이라는 이름의 폴더가 없다면?
if os.path.exists('python') == False:
    #순환적으로 폴더를 생성함 -> exist_ok = True옵션은 이미 존재하더라도 에러 발생 안함 
    os.makedirs('python/test/hello/word', exist_ok=True)    #여러 단계 한번에 생성
    print('python 폴더와 하위 폴더들을 생성 했습니다.')

# 그렇지 않다면?
else : 
    #비어있지 않은 폴더도 강제 삭제 -> 존재하지 않는 폴더인 경우 에러 
    shutil.rmtree('python')
    print('python 폴더가 삭제되었습니다.')


# 이번 코드는 폴더 트리(계층 구조) 생성과 강제 삭제입니다.
#  shutil 모듈
# 파일/폴더 복사
# 폴더 전체 삭제
# 이동
# 압축
# os보다 더 강력한 파일 관리 기능 제공

# exists() 반환값
# 존재 → True
# 없음 → False
```

```python
# 폴더의 이동 
if os.path.exists('python'):
    shutil.move('python', '../created')
    print('python 폴더가 ../created 폴더로 이동되었습니다.')

# shutil.move(src, dst)
# src	원본 경로
# dst	이동할 위치
```

```python
# 폴더의 복사 
if os.path.exists('../created'):
    shutil.copytree('../created', './copy')     #shutil.copytree(src, dst), 폴더 전체를 그대로 복사
    print('python 폴더가 copy로 복사되었습니다.')


# 상위폴더
# └── created
#       └── python

# 현재폴더
# └── copy
#       └── python

# FileExistsError: [WinError 183] 파일이 이미 있으므로 만들 수 없습니다: './copy'
# Output is truncated. View as a scrollable element or open in a text editor. Adjust cell output settings...
```

```python
# os 모듈 내의 path 객체가 갖는 exsists() 함수를 사용 
# -> "hello.txt"라는 파일이 존재하지 않는다면? 
if os.path.exists('hello.txt') == False: 
    #테스트용 파일 생성 
    with open ("hello.txt", "w", encoding="utf-8") as f:    #open(파일명, 모드, 인코딩)
        f.write("Life is too short, you need python")
        print('hello.txt 파일을 생성했습니다.')

    # 생성한 파일을 복사함 -> 이미 존재할 경우 덮어씀 
    shutil.copy('hello.txt', 'world.txt')                   # copy(src, dst)
    print('hello.txt가 world.txt로 복사되었습니다.')

# 그렇지 않다면? 
# -> "hello.txt"라는 파일이 존재한다면?
else: 
    os.remove('hello.txt')                  # 파일 삭제 , 존재하지 않으면 에러 발생 
    print('hello.txt가 삭제되었습니다.')
    os.remove('world.txt')
    print('world.txt가 삭제되었습니다.')


# 1. hello.txt 존재 여부 확인
# 2. 없으면:
#      파일 생성
#      파일 복사
# 3. 있으면:
#      파일 삭제

# 실행 흐름
# hello.txt 없음
# open() 실행
# 파일 생성
# write()로 내용 기록
# 블록 종료
# 자동 close()
```

    - 형식이 갖춰진 문자열을 생성한 후 리턴한다.

    - datetime 객체를 → 지정한 형식의 문자열로 변환

- %M	분	08
- %S	초	32

### 두 날자 간의 차이 계산하기

```python
"""환경 준비 (모듈 import)
메일 정보 정의 (설정값 세팅)
메일 객체 생성
본문 + 첨부파일 구성
SMTP 서버 연결
인증 → 전송 → 종료


os → 파일 존재 확인
SMTP → 메일 서버 통신
MIMEText → 본문 생성
MIMEApplication → 파일 첨부 변환
MIMEMultipart → 본문 + 첨부를 하나로 묶는 컨테이너

여기서 중요한 건:
“메일은 단순 문자열이 아니라, 여러 파트를 가진 구조화된 데이터”
그래서 Multipart가 필요합니다."""
```

```python
# -> 경로 정보를 취득하기 위한 모듈 
import os 
# -> 발송서버와 연동하기 위한 모듈 
from smtplib import SMTP 
# -> 본문 구성 기능 
from email.mime.text import MIMEText 
# -> 파일을 Multipart 형식으로 변환 
from email.mime.application import MIMEApplication
# -> 파일을 본문에 추가하는 기능 제공 
from email.mime.multipart import MIMEMultipart
```

```python
# 보내는 사람 주소 (로그인 계정과 동일 해야 함)
from_addr = "asskjj123@naver.com"

# 받는 사람 주소 
to_addr =" @gmail.com"

# 제목
subject = "파이선 메일 발송 테스트 ~!!! 최연주 입니다(´▽`ʃ♡ƪ)"

# 본문내용 작성 -> 삼중따옴표 안에서는 줄바꿈이 자유롭다. 
content = """안녕하세요 ~ 파이썬 발송하는 메일 입니다 ~!!! 
잘 발송되는지 확인해보도록 하겠습니다 ! 그동안 감사했습니다. 
주영아 강사님 건강하세요 ✧ლ(╹◡╹ლ) 새해 복 많이 받으세요 ~!!! """

# 첨부파일이 필요 없는 경우 
# files = []

# 첨부파일이 필요한 경우 (파일이 실제로 존재해야 함)
files = ["hello.txt", "world.txt"]



"""발신자 주소는 로그인 계정과 동일해야 함
수신자 여러 명 가능 (리스트로 바꿀 수 있음)
files는 리스트 구조 → 반복문으로 처리하기 위함"""
```

```python
# 컨텐츠 형식 (plain or html)
content_type = "plain"

# 로그인 계정 이름 (네이버=아이디, 구글=메일주소)
username = "asskjj123"

# 비밀번호 (네이버=개인비밀번호, 애플리케이션 비밀번호, 구글=앱 비밀번호)
password = "WVG5YCS1QK5G"

# 구글 발송 서버 주소와 포트 (고정값)
# smtp = "smtp.gmail.com"
# port = 587

# 네이버 발송 서버 주소오 포트 (고정값)
smtp = "smtp.naver.com"
port = 465
```

```python
# 메일 발송 정보를 저장하기 위한 객체 
msg = MIMEMultipart()

# 헤더 설정 
# 이건 메일의 메타정보
# SMTP는 결국 이 헤더를 서버로 보냅니다.
msg['Subject'] = subject #메일 제목 
msg['From'] = from_addr # 보내는 사람 
msg['To'] = to_addr     # 받는 사람 

# 본문 설정 -> 메일의 내용과 형식 지정 
# MIMEText는 “본문을 메일 형식으로 변환”하는 것.
# content_type이: plain → 일반 텍스트
# html → HTML 메일
msg.attach(MIMEText(content, content_type))


"""
메일
 ├─ 헤더
 │   ├─ 제목
 │   ├─ 보낸 사람
 │   └─ 받는 사람
 ├─ 본문
 └─ 첨부파일
 """
```

```python
# 리스트 변수의 원소가 하나라도 존재할 경우 True 
# 이건 리스트가 비어있지 않으면 True.
if files:
    for file_item in files :
        if os.path.exists(file_item):
            # 바이너리(b) 형식으로 읽기(r)
            with open(file_item, 'rb') as f:    # 파일은 이미지/텍스트 구분 없이 전송되므로 바이너리로 읽어야 깨지지 않습니다.
                # 전체 경로에서 파일의 이름만 추출 
                basename = os.path.basename(file_item)  # basename 추출 이유, 경로 제거하고 파일 이름만 추출
                # 파일의 내용과 파일이름을 메일에 첨부할 형식으로 변환 
                part = MIMEApplication(f.read(), Name=basename)

                # 파일 첨부 
                part['Content-Disposition'] = 'attachment; filename="%s"' % basename 
                msg.attach(part)

                print(basename, "(이)가 첨부되었습니다.")

"""
파일 목록이 있는가?
    → 있다
        → 파일이 실제 존재하는가?
            → 있다
                → 바이너리로 읽기
                → MIME 형식으로 변환
                → 메일에 attach
"""
```

```python
mail = SMTP(smtp)
# 메일 서버 접속 
mail.ehlo()
# 메일 서버 연동 설정 
mail.starttls()
# 메일 서버 로그인 
mail.login(username, password)
# 메일 보내기 
mail.sendmail(from_addr, to_addr, msg.as_string() )
# 메일 서버 접속 종료 
mail.quit()
```
