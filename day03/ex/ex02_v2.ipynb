#### selenium 형식

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

# 1. 크롬 옵션 설정
options = Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(options=options)

# 2. 페이지 접속
url = "여기에_크롤링할_URL_입력"
driver.get(url)

time.sleep(3)

# ----------------------------------
# 3. 무한 스크롤 구현
# ----------------------------------

prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 스크롤 맨 아래로 이동
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    # 현재 높이 확인
    curr_height = driver.execute_script("return document.body.scrollHeight")

    # 더 이상 늘어나지 않으면 종료
    if curr_height == prev_height:
        break

    prev_height = curr_height

print("스크롤 완료")

# ----------------------------------
# 4. 제목 요소들 찾기
# ----------------------------------

titles = driver.find_elements(By.CLASS_NAME, "sds-comps-text-type-headline1")

title_list = []

for t in titles:
    text = t.text.strip()
    if text != "":
        title_list.append(text)

# ----------------------------------
# 5. 결과 확인
# ----------------------------------

for title in title_list:
    print(title)

driver.quit()
```

#### selenium 형식

```python
# 1. 필요한 모듈
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

# 2. 크롬 드라이버 실행
driver = webdriver.Chrome()

# 3. 네이버 검색 페이지 접속 (블로그 탭)
keyword = "빅데이터"
url = f"https://search.naver.com/search.naver?query={keyword}&where=blog"

driver.get(url)
time.sleep(3)

# 4. 무한 스크롤 처리
prev_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

    curr_height = driver.execute_script("return document.body.scrollHeight")

    if curr_height == prev_height:
        break

    prev_height = curr_height

print("스크롤 완료")

# 5. 페이지 소스 가져오기
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# 6. 요약 텍스트만 추출
summaries = soup.select("span.sds-comps-text-type-body")

for i, summary in enumerate(summaries, 1):
    text = summary.get_text(strip=True)
    print(f"{i}. {text}")

driver.quit()
```

```python
# 지피티 추천
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wait = WebDriverWait(driver, 10)

date_element = wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, "span.sds-comps-profile-info-subtext")
    )
)

date = date_element.text
print(date)
```
