```python
# Step 1. 필요한 모듈 로딩
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# Step 2. 사용자 입력
print("=" * 100)
print(" 네이버 블로그 자료 수집 크롤러입니다.")
print("=" * 100)

query_txt = input("1. 수집할 키워드를 입력하세요 : ")

# Step 3. 크롬 드라이버 실행
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

url = "https://www.naver.com/"
driver.get(url)
time.sleep(3)
driver.maximize_window()

# Step 4. 검색어 입력
search_box = driver.find_element(By.ID, "query")
search_box.click()
search_box.send_keys(query_txt)
search_box.send_keys("\n")
time.sleep(3)

# Step 5. 블로그 탭 클릭
driver.find_element(By.LINK_TEXT, "블로그").click()
time.sleep(3)

# ----------------------------------------
# ⭐ Step 6. 무한 스크롤 처리
# ----------------------------------------

scroll_count = 5   # 스크롤 횟수 (원하는 만큼 조절 가능)

for i in range(scroll_count):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

print("스크롤 완료")

# ----------------------------------------
# Step 7. HTML 가져오기
# ----------------------------------------

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

# ----------------------------------------
# Step 8. 제목 + 요약 추출
# ----------------------------------------

titles = soup.select("span.sds-comps-text-type-headline1")
summaries = soup.select("span.sds-comps-text-type-body")

title_list = []
summary_list = []

for t in titles:
    title_list.append(t.get_text(strip=True))

for s in summaries:
    summary_list.append(s.get_text(strip=True))

# 길이 맞추기 (혹시 다를 경우 대비)
min_len = min(len(title_list), len(summary_list))
title_list = title_list[:min_len]
summary_list = summary_list[:min_len]

# ----------------------------------------
# Step 9. DataFrame 생성
# ----------------------------------------

df = pd.DataFrame({
    "제목": title_list,
    "요약": summary_list
})

# ----------------------------------------
# Step 10. 파일 저장
# ----------------------------------------

fx_name = "naver_blog.xlsx"
fc_name = "naver_blog.csv"

df.to_excel(fx_name, index=False, engine="openpyxl")
df.to_csv(fc_name, index=False, encoding="utf-8-sig")

print("데이터 수집이 완료되었습니다.")
print("총 수집 건수:", len(df))

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

print("=" * 100)
print(" 네이버 블로그 자료 수집 크롤러입니다.")
print("=" * 100)

query_txt = input("1. 수집할 키워드를 입력하세요 : ")

# 크롬 실행
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

# 🔥 블로그 검색 페이지로 바로 이동
url = f"https://search.naver.com/search.naver?where=blog&query={query_txt}"
driver.get(url)
driver.maximize_window()

# 🔥 검색 결과 로딩될 때까지 기다림 (핵심)
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "ul.lst_view")))

print("검색 결과 로딩 완료")

# ------------------------------
# 무한 스크롤
# ------------------------------
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

print("스크롤 완료")

# ------------------------------
# 제목 + 요약 추출
# ------------------------------
titles = driver.find_elements(By.CSS_SELECTOR, "a.api_txt_lines.total_tit")
summaries = driver.find_elements(By.CSS_SELECTOR, "div.api_txt_lines.dsc_txt")

title_list = [t.text for t in titles if t.text.strip() != ""]
summary_list = [s.text for s in summaries if s.text.strip() != ""]

min_len = min(len(title_list), len(summary_list))

df = pd.DataFrame({
    "제목": title_list[:min_len],
    "요약": summary_list[:min_len]
})

print("총 수집 건수:", len(df))

df.to_excel("naver_blog.xlsx", index=False, engine="openpyxl")
df.to_csv("naver_blog.csv", index=False, encoding="utf-8-sig")

print("데이터 수집이 완료되었습니다.")

driver.quit()
```
