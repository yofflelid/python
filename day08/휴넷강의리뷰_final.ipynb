```python
# 이거 됨 

# Step 6. 현재 검색 결과 페이지 HTML 가져오기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

course_list = soup.find('ul', id='divEducationPartialList')
course_items = course_list.find_all('li')

print("\n")
print("=" * 100)
print("검색된 강의 목록입니다.")
print("=" * 100)

title2 = []
img_src2 = []
detail_url2 = []
score2 = []
review_num2 = []
price2 = []

for idx, li in enumerate(course_items, start=1):
    img_src = li.find('img')['src']
    title = li.find('div', class_='title').get_text(strip=True)

    href_text = li.find('div', class_='title').find('a')['href']
    detail_url = href_text.split("'")[1]

    score = li.find('span', class_='ex').find('strong', class_='emp').get_text(strip=True)
    review_num = li.find('span', class_='review').find('strong', class_='emp').get_text(strip=True)
    price = li.find('span', class_='expense').find('strong', class_='emp').get_text(strip=True)

    title2.append(title)
    img_src2.append(img_src)
    detail_url2.append(detail_url)
    score2.append(score)
    review_num2.append(review_num)
    price2.append(price)

    print("%s. 강의명: %s" % (idx, title))
    print("   이미지:", img_src)
    print("   만족도:", score)
    print("   후기 수:", review_num)
    print("   금액:", price)
    print("   상세 URL:", detail_url)
    print("-" * 100)

# Step 7. 몇 번째 강의의 리뷰를 볼지 입력받기
course_no = int(input("\n4. 몇 번째 강의의 리뷰를 보시겠습니까?: "))

selected_title = title2[course_no - 1]
selected_url = detail_url2[course_no - 1]

print("\n선택한 강의명:", selected_title)
print("상세페이지 URL:", selected_url)

# Step 8. 선택한 강의 상세페이지로 이동
driver.get(selected_url)
time.sleep(3)

# Step 9. 학습후기 탭 클릭
driver.find_element(By.ID, "tab3").click()
time.sleep(3)

# Step 10. 리뷰 페이지 HTML 다시 읽기
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

total_review = soup.find('strong', id='spanReviewCnt').get_text(strip=True)
total_review = int(total_review)

print("\n총 리뷰 수는 %s건 입니다." % total_review)

# Step 11. 수집할 리뷰 수 입력받기
review_cnt_input = input("5. 수집할 리뷰는 총 몇건입니까?(엔터 입력 시 전체 수집): ")

if review_cnt_input == '':
    review_cnt = total_review
    print("전체 리뷰 %s건을 수집합니다." % review_cnt)
else:
    review_cnt = int(review_cnt_input)
    if review_cnt > total_review:
        review_cnt = total_review
        print("입력한 수가 전체 리뷰 수보다 많아서 %s건으로 조정합니다." % review_cnt)

page_cnt = math.ceil(review_cnt / 10)
print("총 %s 페이지를 수집합니다." % page_cnt)

# Step 12. 저장할 리스트 생성
writer_name2 = []
writer_id2 = []
review_text2 = []

# txt 저장 준비
n = time.localtime()
s = '%04d-%02d-%02d-%02d-%02d-%02d' % (n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)

save_dir = f_dir + s + '-' + query_txt
os.makedirs(save_dir, exist_ok=True)

ff_name = save_dir + '\\' + s + '-' + query_txt + '.txt'
fc_name = save_dir + '\\' + s + '-' + query_txt + '.csv'
fx_name = save_dir + '\\' + s + '-' + query_txt + '.xlsx'

f = open(ff_name, 'a', encoding='UTF-8')

# Step 13. 리뷰 수집
count = 0

for page in range(page_cnt):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    review_area = soup.find('div', id='reviewList')
    review_ul = review_area.find('ul', class_='view-epilogue-list')
    review_items = review_ul.find_all('li')

    for li in review_items:
        count += 1

        author_info = li.find('div', class_='author')
        author_strong = author_info.find_all('strong')

        writer_name = author_strong[0].get_text(strip=True)
        writer_id = author_strong[1].get_text(strip=True)

        author_info.extract()
        review_text = li.get_text(" ", strip=True)

        print("\n총 %s건 중 %s번째 리뷰 수집 중입니다 =========" % (review_cnt, count))
        print("1. 작성자 이름:", writer_name)
        print("2. 작성자 ID:", writer_id)
        print("3. 리뷰 내용:", review_text)

        f.write("\n")
        f.write("총 %s건 중 %s번째 리뷰 수집 중입니다 =========\n" % (review_cnt, count))
        f.write("1. 작성자 이름: " + writer_name + "\n")
        f.write("2. 작성자 ID: " + writer_id + "\n")
        f.write("3. 리뷰 내용: " + review_text + "\n")

        writer_name2.append(writer_name)
        writer_id2.append(writer_id)
        review_text2.append(review_text)

        time.sleep(0.2)

        if count == review_cnt:
            break

    if count == review_cnt:
        break

    # 다음 페이지 클릭
    try:
        driver.find_element(By.CSS_SELECTOR, 'a.next').click()
        time.sleep(2)
    except:
        print("다음 페이지가 없어 수집을 종료합니다.")
        break

f.close()

# Step 14. 결과 저장
import pandas as pd

result = pd.DataFrame()
result['강의명'] = [selected_title] * len(writer_name2)
result['작성자이름'] = writer_name2
result['작성자ID'] = writer_id2
result['리뷰내용'] = review_text2

result.to_csv(fc_name, index=False, encoding='utf-8-sig')
result.to_excel(fx_name, index=False)

# Step 15. 요약 출력
print("\n" + "=" * 100)
print("1. 요청된 총 %s 건의 리뷰 중에서 실제 수집된 리뷰 수는 %s 건입니다." % (review_cnt, len(writer_name2)))
print("2. txt 파일 저장 완료:", ff_name)
print("3. csv 파일 저장 완료:", fc_name)
print("4. xlsx 파일 저장 완료:", fx_name)
print("=" * 100)

driver.close()
```
