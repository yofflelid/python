```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time
import math
import os


# Step 1. 사용자 입력
print("=" * 100)
print("휴넷 강의 리뷰 수집기")
print("=" * 100)

query_txt = input("1. 검색할 키워드를 입력하세요: ")

f_dir = input("2. 파일 저장 경로를 입력하세요(예: c:\\py_temp\\): ")
if f_dir == "":
    f_dir = "c:\\py_temp\\"

# Step 2. 드라이버 실행
s = Service("c:/py_temp/chromedriver.exe")
driver = webdriver.Chrome(service=s)

driver.get("https://www.hunet.co.kr/")
driver.maximize_window()
time.sleep(3)

# Step 3. 검색
element = driver.find_element(By.ID, "txtKeyword")
element.click()
element.send_keys(query_txt)
element.send_keys(Keys.ENTER)
time.sleep(3)

# Step 4. 현재 검색 결과 HTML 읽기
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

course_list = soup.find("ul", id="divEducationPartialList")
course_items = course_list.find_all("li")

# Step 5. 강의 목록 출력 + 정보 저장
title_list = []
img_list = []
url_list = []
score_list = []
review_cnt_list = []
price_list = []

print("\n[검색된 강의 목록]")
print("=" * 100)

for idx, li in enumerate(course_items, start=1):
    title = li.find("div", class_="title").get_text(strip=True)
    img_src = li.find("img")["src"]

    href_text = li.find("div", class_="title").find("a")["href"]
    detail_url = href_text.split("'")[1]

    score = li.find("span", class_="ex").find("strong").get_text(strip=True)
    review_cnt = li.find("span", class_="review").find("strong").get_text(strip=True)
    price = li.find("span", class_="expense").find("strong").get_text(strip=True)

    title_list.append(title)
    img_list.append(img_src)
    url_list.append(detail_url)
    score_list.append(score)
    review_cnt_list.append(review_cnt)
    price_list.append(price)

    print(f"{idx}. {title}")
    print(f"   만족도: {score} / 후기: {review_cnt} / 금액: {price}")
    print("-" * 100)

# Step 6. 몇 번째 강의 리뷰를 볼지 선택
course_no = int(input("\n3. 몇 번째 강의의 리뷰를 보시겠습니까?: "))

selected_title = title_list[course_no - 1]
selected_url = url_list[course_no - 1]

print("\n선택한 강의:", selected_title)
print("상세 URL:", selected_url)

# Step 7. 상세페이지 이동
driver.get(selected_url)
time.sleep(3)

# Step 8. 학습후기 탭 클릭
driver.find_element(By.ID, "tab3").click()
time.sleep(3)

# Step 9. 리뷰 페이지 HTML 다시 읽기
html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

total_review = soup.find("strong", id="spanReviewCnt").get_text(strip=True)
total_review = int(total_review)

print(f"\n총 리뷰 수: {total_review}건")

# Step 10. 수집할 리뷰 수 입력
cnt = int(input("4. 수집할 리뷰 수는 몇 건입니까?: "))

if cnt > total_review:
    cnt = total_review
    print(f"전체 리뷰 수보다 커서 {total_review}건으로 조정합니다.")

page_cnt = math.ceil(cnt / 10)
print("수집할 페이지 수:", page_cnt)

# Step 11. 리뷰 데이터 저장용 리스트
writer_name_list = []
writer_id_list = []
review_text_list = []

count = 0

# Step 12. 페이지 반복
for page in range(page_cnt):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    review_area = soup.find("div", id="reviewList")
    review_ul = review_area.find("ul", class_="view-epilogue-list")
    review_items = review_ul.find_all("li")

    for li in review_items:
        author_info = li.find("div", class_="author")
        author_strong = author_info.find_all("strong")

        writer_name = author_strong[0].get_text(strip=True)
        writer_id = author_strong[1].get_text(strip=True)

        author_info.extract()
        review_text = li.get_text(" ", strip=True)

        count += 1

        print(f"\n총 {cnt}건 중 {count}번째 리뷰 수집 중")
        print("작성자 이름:", writer_name)
        print("작성자 ID:", writer_id)
        print("리뷰 내용:", review_text)

        writer_name_list.append(writer_name)
        writer_id_list.append(writer_id)
        review_text_list.append(review_text)

        if count == cnt:
            break

    if count == cnt:
        break

    # 다음 페이지 클릭
    try:
        driver.find_element(By.CSS_SELECTOR, "a.next").click()
        time.sleep(2)
    except:
        print("다음 페이지가 없어 수집을 종료합니다.")
        break

# Step 13. 데이터프레임 만들기
df = pd.DataFrame({
    "강의명": [selected_title] * len(writer_name_list),
    "작성자이름": writer_name_list,
    "작성자ID": writer_id_list,
    "리뷰내용": review_text_list
})

# Step 14. 파일 저장
n = time.localtime()
s = "%04d-%02d-%02d-%02d-%02d-%02d" % (
    n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec
)

save_dir = os.path.join(f_dir, s + "-" + query_txt)
os.makedirs(save_dir, exist_ok=True)

txt_name = os.path.join(save_dir, s + "-" + query_txt + ".txt")
csv_name = os.path.join(save_dir, s + "-" + query_txt + ".csv")
xlsx_name = os.path.join(save_dir, s + "-" + query_txt + ".xlsx")

# txt 저장
with open(txt_name, "w", encoding="utf-8") as f:
    for i in range(len(writer_name_list)):
        f.write(f"{i+1}번째 리뷰\n")
        f.write("강의명: " + selected_title + "\n")
        f.write("작성자 이름: " + writer_name_list[i] + "\n")
        f.write("작성자 ID: " + writer_id_list[i] + "\n")
        f.write("리뷰 내용: " + review_text_list[i] + "\n")
        f.write("=" * 80 + "\n")

# csv, xlsx 저장
df.to_csv(csv_name, index=False, encoding="utf-8-sig")
df.to_excel(xlsx_name, index=False)

print("\n" + "=" * 100)
print("리뷰 수집 완료")
print("txt 저장:", txt_name)
print("csv 저장:", csv_name)
print("xlsx 저장:", xlsx_name)
print("=" * 100)

driver.quit()
```
