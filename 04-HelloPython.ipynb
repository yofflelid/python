### 4. 문자열 포메팅
- 특정 값이 치환될 형식 문자를 포함한 문자열의 형식(format)을 미리 정해놓고  
    - 형식 문자에 특정 변수 값을 적용하여 완성된 문자열을 만드는 방법 
- 문자열에 포함될 수 있는 형식 문자 
    - %d %f %s %o %x %% 
- 포매팅의 동작 원리 
    - **"문자열 안의 형식문자" % 변수**
    - 구조 분해해보면
    - 문자열 안에 %d 같은 형식 문자 존재
    - % 뒤에 변수 값을 전달
    - 형식 문자가 그 값을 해당 타입으로 변환해서 삽입

- 사용방법 
    - 형식 문자를 포함한 문자열 뒤에 % 연산자를 적용하여 치환할 값을 명시한다. 
    - 직접 출력
    - 두 개 이상의 형식 문자를 포함된 문자열에 치환할 값들은 소괄호()로 묶어 콤마로 구분한다. 
        - 형식 문자가 두 개 이상인 경우 치환할 값이나 변수를 괄호로 묶는다. 

- 정수의 자릿수 표현 **%자리수d**     /   %[옵션][폭][타입]
    - 출력되는 문자열을 자릿수에 맞추어 *오른쪽으로 정렬* 시킨다.
    - 지정된 자리수를 초과하면 그냥 그대로 출력됨 (잘리지 않음)
    - 마이너스(-)는 왼쪽 정렬
    - 문자열은 숫자가 아니기 때문에 "0으로 채운다"는 개념이 없음.
        - 그래서 %010s은 지원되지 않거나 의미가 없음. /  %10s ~ 빈자리에 공백 

- 실수의 자릿수 표현(1) 
    - 형식 문자에 숫자를 포함하여 자릿수 지정 
        - 출력할 값이 지정된 소수점 자릿수를 초과할 경우 반올림 된다. 

- 인덱스 번호를 사용한 문자열 포메팅 ~ 객체이름.format
    - 인덱스를 갖는 치환 문자를 포함한 문자열 객체에 대한 메서드 호출 
    - {숫자} 형식의 치환문자를 문자열에 포함한다. 
        - "과자는 {0}{1}원 이다." 
    - {이름} 형식의 치환문자를 사용할 수 있다. 
        - "과자는 {price}{unit} 입니다." 
        - unit="원", price=2000 
    - 인덱스 번호와 이름의 혼합사용 
        - 숫자형식의 치환문자가 format() 메서드 안에서 먼저 위치해야 한다. 
        - 인덱스 번호는 format() 함수 내의 순서 ~ 형식 문자열 안에서는 자유롭게 배치 가능

```python
# 원소의 수를 조회한 다음 직접 반복 범위 지정하기
# -> 단점: 각 원소가 저장되는 변수가 없기 때문에 
#           원소에 대한 접근을 리스트를 통해 직접 수행 해야 한다. 
seasons = ['봄', '여름', '가을', '겨울']

size = len(seasons) 
for i in range(0, size):
    print(i , "번째 원소", seasons[i]) 
```

```python
# 리스트의 원소를 for문에서 설정한 변수에 직접 할당한다. 

msg = ['python', 'is', 'good'] 

# -> 리스트의 각 원소를 순차적으로 i에 대입하면서 반복
# -> 단점: 몇 번째 원소인지는 알 수 없다. 
for i in msg :
    print(i) 
```

```python
#리스트 원소에 대한 합계, 평균 구하기 

# 리스트 원소의 총점, 평균 구하기 
# 총합을 위해서는 반복 시작 전에 합계를 저장할 변수를 0으로 초기화 해야 한다. 

# 성적에 대한 총점과 평균 구하기 
score = [100, 82, 98, 76, 51]

# -> 총점을 저장할 변수 
total = 0 

for v in score : 
    # 총점은 각 원소를 누적해서 더한다. 
    total += v

# 평균은 총 합에서 원소의 수를 나눈다.
# -> "//" 는 나눗셈에서 정수부분의 몫만 구한다. 
avg = total // len(score)

print("총첨", total, ", 평균", avg)
```

```python
# 리스트의 인덱스와 원소 탐색하기 

# i에는 리스트의 인덱스가 저장된다. 
# v에는 리스트의 값(원소)가 저장된다. 
# i, v는 단순 변수 이므로 개발자 편의에 따라 이름지을 수 있다. 
city = ['서울', '인천', '대구', '대전', '광주', '부산']

for i , v in enumerate(city):
    print(i, "번째 값", v)
```

```python
# 2차원 리스트 탐색 패턴(1) - step 1 

# 2차원 리스트의 원소 출력하기 
# 2차원 리스트는 각 원소가 각각 개별적인 리스트인 경우 이므로 
# 2차원 리스트의 특정 원소를 출력하면 
# 원소로 포함되어 있는 개별 리스트를 얻을 수 있다. 

mydata = [[88, 72, 83, 90], [63, 77, 72, 80]]

size = len(mydata)
print("size = ", size) 

for i in range(0, size) :
    #
    print(i, "번째 데이터->", mydata[i])
    
```

```python
# 2차원 리스트 탐색 패턴(2) - step 2

# 2차원 리스트의 각 원소에 대한 반복문 구성 

mydata = [[88, 72, 83, 90], [63, 77, 72, 80]]

size = len(mydata)              #2차원 리스트의 길이는 행의 수를 반환한다. 
print("size = ", size) 

for i in range(0, size) :       # 행의 수 만큼 반복하는 구성 
    # i번째 행에 포함되어 있는 개별 리스트를 출력한다. 
    print(i, "번째 데이터->", mydata[i])

    total = 0                   # 각 행 단위 합계 이므로 매 행마다 합계 변수를 새로 초기화 한다.
    size = len(mydata[i])       # 각 행에 포함된 개별 리스트의 길이 

    for j in range (0, size) :  # mydata[i]에 대한 새로운 반복문 구성 
        total += mydata[i][j]

    average = total / size      # 행 단위 평균값 

    print(i, "번째 데이터 ==> 총합", total, "평균 = ", average) 
```

```python
# 2차원 리스트 탐색 패턴(2) - step 1 

# 리스트를 반복문에 직접 적용한 경우 
# 반복문의 변수가 리스트의 원소가 된다. 
# 2차원 리스트의 경우 원소는 각각의 자식 리스트가 된다. 
 
mydata = [ [88, 72, 83, 90], [63, 77, 72, 80]]

for row in mydata :             # 행의 수 만큼 반복하는 구성 
    # mydata의 원소 하나인 리스트를 출력한다. 몇 번째 항목인지는 알 수 없다. 
    print( "이번에 출력되는 데이터 ->", row)
```

```python
# 2차원 리스트 탐색 패턴(2) - step 2
# 행 단위 합계, 평균 

mydata = [ [88, 72, 83, 90], [63, 77, 72, 80]]

for row in mydata :             # 행의 수 만큼 반복하는 구성 
    # mydata의 원소 하나인 리스트를 출력한다. 몇 번째 항목인지는 알 수 없다. 
    print( "이번에 출력되는 데이터 -> ", row)

    total = 0                   # 각 행 단위 합계 이므로 매 행마다 합계 변수를 새로 초기화 한다.

    for col in row :            # row에 대한 새로운 반복문 구성 
        total += col 

    average = total / len(row)  # 행 단위 평균값 

    print( row, "에 대한 총합", total, ", 평균 =", average)
```

```python
# 대 소문자 여부 검사하기 

# 대문자 검사 -> 문자열객체.isupper(), 소문자 검사 -> 문자열객체.islower()

# 문자열이 대문자로 구성된 경우 True 
test3 = "HELLO"
print(test3.isupper())

# 문자열이 대문자로 구성되지 않은 경우 False
test4 = "Hello"
print(test4.isupper())
```

```python
# 문자열이 소문자로 구성된 경우 True 
test5 = "hello"
print(test5.islower())

# 문자열이 소문자로 구성되지 않은 경우 False
test6 = "Hello"
print(test6.islower())
```

```python
# 대소문자 변환하기 (1)

# upper(), lower() 메서드의 사용 

# 실습을 위한 문자열 
mystr = "Hello Python"
```

```python
# 공백 지우기 

# 문자열객체.strip() , 문자열객체.lstrip(), 문자열객체.rstrip() 
# 문자열의 앞뒤 공백, 왼쪽 공백, 오른쪽 공백을 삭제한다. 

# 실습을 위한 문자열 ( 앞 , 뒤에 공백이 포함되어 있다. )
k = "          python          "
print( "["   +   k   +    "]" )

# 1) 앞 뒤 공백 제거 
x = k.strip()
print( "["   +   x   +  "]" )

# 2) 왼쪽 공백 제거 
y = k.lstrip() 
print( "["    + y + "]"    )

# 3) 오른쪽 공백 제거 
z = k.rstrip() 
print( "["  +   z   +   "]")
```

```python
# 문자열 바꾸기 

# 문자열객체.replace("원본값", "변경할 값")
# 문자열에서 원본값을 모두 변경할 값으로 치환한 결과를 리턴한다. (원본은 변하지 않는다.)

#
myMessage = "Hello World"


# 변수.replace( A, B) , 변수에서 A를 B로 변경한 결과를 반환 
newMessage = myMessage.replace("World", "Python")
print( newMessage)

# 원본은 변화 없음 
print(myMessage)
```

```python
# 문자열 자르기  

# 문자열객체.split(기준값)
# 파라미터로 전달된 기준값을 사용하여, 문자열을 잘라낸 후 리스트로 변환한다. 

#AttributeError: 'str' object has no attribute 'slist'

# 문자열을 잘라서 리스트로 변환하기 
sample1 = "파이썬,데이터분석,머신러닝"
slist1 = sample1.split(",") # 문자열을 잘라낼 기준값을 파라미터로 지정 
print(slist1) 
```

```python
# split() 메서드에 파라미터가 없을 경우 공백을 기준으로 잘라낸다. 
sample2 = "파이썬 데이터분석 머신러닝"
slist2 = sample2.split()
print(slist2)
```
