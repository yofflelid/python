```python
for j in full_detail_url :    
    driver.get(j)
    
    time.sleep(5)
    
    html3 = driver.page_source
    soup3 = BeautifulSoup(html3, 'html.parser')
    
    # 해당 페이지의 이미지 수집
    #Step 3. 결과를 저장할 폴더명과 파일명을 설정하고 폴더를 생성합니다.
    import urllib.request
    
    query_txt2 = '인스타그램'
    now = time.localtime()
    s = '%04d-%02d-%02d-%02d-%02d-%02d' % (now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)
    img_dir = f_dir+s+'-'+query_txt2+'-'+query_txt
    os.makedirs(img_dir)
    f_name=f_dir+s+'-'+query_txt2+'-'+query_txt+'\\'+s+'-'+query_txt+'.txt'
    
    file_no = 1
    
    imgs = soup3.find('div','_aagu')
    try :
        img_src1=imgs.find('img')['src']
    except :
        pass
    else :
        img_name = img_dir +'\\'+ str(file_no)+'.jpg'
        #print(img_src1)
        urllib.request.urlretrieve(img_src1,img_name)       
        time.sleep(1)
        
        print("현재 %s 번째 이미지 저장 완료했습니다=======" %file_no)
        print("\n")
    
        file_no += 1
    
    # 해당 페이지의 태그 수집 후 저장
    f = open(f_name, 'a',encoding='UTF-8')
    # 해당 페이지의 태그 수집
    tags_1 = soup3.find('div','x4h1yfo').get_text()
    
    tags_2 = tags_1.translate(bmp_map)
    tags_3 = unicodedata.normalize('NFC', tags_2)
    print('텍스트추출결과:' , tags_3)
    
    for i in tags_3 :
        if i[0:1]=='#' :
            #hash_txt.append(tags_2)
            print(i)
            try:
                f.write("\n" + str(tags_2),'a',encoding='UTF-8')
            except :
                pass
    f.close()
```
